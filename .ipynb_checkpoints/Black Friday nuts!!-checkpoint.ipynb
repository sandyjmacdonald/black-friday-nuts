## Notes

The blue bars and line are a binned histogram and kernel density estimation respectively of the crowd-sourced guesses.

The red bars and line are are a binned histogram and kernel density estimation of a normal distribution with mean of 5,386 (the actual number of nuts) and standard deviation of 5386/2 (a guesstimation!)

In the end, the mean guess came out as 5,268 compared with the actual number of 5,386, just 118 out (just 2.19% error!)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats.mstats import gmean, hmean
# plt.style.use('ggplot')
%matplotlib inline
```

```python
actual = 5386
df_twit = pd.read_csv('nuts-twit.txt', sep='\t', header=None)
df_fb = pd.read_csv('nuts-fb.txt', sep='\t', header=None)
df_ig = pd.read_csv('nuts-ig.txt', sep='\t', header=None)
df = pd.concat([df_twit, df_fb, df_ig])
```

## Facebook

```python
sns.set(style="white", palette="muted", color_codes=True)
normdist = np.random.normal(actual, actual/2.0, len(df_fb))
fig, ax = plt.subplots()
sns.distplot(df_fb, color="b", bins=None, ax=ax)
sns.distplot(normdist, color="r", bins=None, ax=ax)
plt.axvline(x=np.average(normdist), color="b", linestyle="dashed")
plt.axvline(x=np.average(df_fb), color="r", linestyle="dashed")
plt.tight_layout()
```

## Twitter

```python
sns.set(style="white", palette="muted", color_codes=True)
normdist = np.random.normal(actual, actual/2.0, len(df_twit))
fig, ax = plt.subplots()
sns.distplot(df_twit, color="b", bins=None, ax=ax)
sns.distplot(normdist, color="r", bins=None, ax=ax)
plt.axvline(x=np.average(normdist), color="b", linestyle="dashed")
plt.axvline(x=np.average(df_twit), color="r", linestyle="dashed")
plt.tight_layout()
```

## Instagram

```python
sns.set(style="white", palette="muted", color_codes=True)
normdist = np.random.normal(actual, actual/2.0, len(df_ig))
fig, ax = plt.subplots()
sns.distplot(df_ig, color="b", bins=None, ax=ax)
sns.distplot(normdist, color="r", bins=None, ax=ax)
plt.axvline(x=np.average(normdist), color="b", linestyle="dashed")
plt.axvline(x=np.average(df_ig), color="r", linestyle="dashed")
plt.tight_layout()
```

## Facebook, Twitter, Instagram combined

```python
sns.set(style="white", palette="muted", color_codes=True)
normdist = np.random.normal(actual, actual/2.0, len(df))
fig, ax = plt.subplots()
sns.distplot(df, color="b", bins=None, ax=ax)
sns.distplot(normdist, color="r", bins=None, ax=ax)
plt.axvline(x=np.average(normdist), color="b", linestyle="dashed")
plt.axvline(x=np.average(df), color="r", linestyle="dashed")
plt.tight_layout()
```

```python
sns.set(style="white", palette="muted", color_codes=True)
normdist = np.random.normal(actual, actual/2.0, len(df))
fig, ax = plt.subplots()
sns.distplot(df, color="b", bins=None, ax=ax)
sns.distplot(normdist, color="r", bins=None, ax=ax)
sns.despine(left=True)
ax.set_yticks([])
ax.set_xticks([])
plt.axvline(x=np.average(normdist), color="b", linestyle="dashed")
plt.axvline(x=np.average(df), color="r", linestyle="dashed")
plt.tight_layout()
```

```python
avg = np.average(df[0])
med = df.median()
g_mean = gmean(df[0])
h_mean = hmean(df[0])
std = df.std()

print 'actual: %i\nmean: %f\nmedian: %f\ngeom. mean: %f\nharm. mean: %f\nstd. dev.: %f' % (actual, avg, med, g_mean, h_mean, std)
```
